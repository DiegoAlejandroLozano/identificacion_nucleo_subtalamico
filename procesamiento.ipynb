## Librerias necesarias

```python
import scipy.io
import pandas as pd
import os
```

### 1. Procesamiento de los registros Stn de los pacientes 3, 4, 5 y 6.

Función encargada de procesar los datos

```python
def procesar_registros(ruta:str) -> pd.DataFrame:
    datos_25000 = scipy.io.loadmat(ruta)['sig']
    datos_2500 = datos_25000.reshape((10,2500))
    return pd.DataFrame(datos_2500)
```

Se crea el Dataframe que contendrá los registros modificados de las muestras del STN

```python
df_stn = pd.DataFrame()
```

Se recorren todos los registros obtenidos del núcleo STN

```python
for i in range(3,7):
    archivos = 'BD_original/DB_UTP/Paciente {}/Stn/'.format(i)
    for archivo in os.listdir(archivos):
        ruta = "BD_original/DB_UTP/Paciente {}/Stn/{}".format(i, archivo)
        registro_procesado = procesar_registros(ruta=ruta)
        df_stn = pd.concat([df_stn, registro_procesado], ignore_index=True)
df_stn    
```

"Se añaden nombres a las columnas del DataFrame y se introduce una nueva columna que contendrá una etiqueta para identificar si los registros corresponden al núcleo subtalámico. Para señalar que un registro pertenece a dicho núcleo, se asignará el valor 1 en la columna denominada 'estructura'."

```python
#Se agregan el nombre a la columnas
columnas = []
for i in range(1,2501):
    columnas.append("muestra {}".format(i))
df_stn.columns = columnas

#Se agrega la nueva columna con la etiqueta
df_stn['estructura'] = 1

df_stn
```

### 2. Procesamiento de los registros de las otras estructuras de los pacientes 3, 4, 5 y 6.

Se crea el Dataframe que contendrá los registros modificados de las muestras de las otras estructuras

```python
df_otras = pd.DataFrame()
```

Se recorren todos los registros de las otras estructuras cerebrales

```python
for i in range(3,7):
    archivos = 'BD_original/DB_UTP/Paciente {}/Otras Zonas/'.format(i)
    for archivo in os.listdir(archivos):
        ruta = "BD_original/DB_UTP/Paciente {}/Otras Zonas/{}".format(i, archivo)
        registro_procesado = procesar_registros(ruta=ruta)
        df_otras = pd.concat([df_otras, registro_procesado], ignore_index=True)
df_otras
```

Siguiendo el mismo proceso que en el DataFrame de los registros del núcleo subtalámico, se asignarán nombres a las columnas y se añadirá una nueva columna para etiquetar los registros correspondientes a otras regiones cerebrales. La nueva columna contendrá el valor 0 para distinguir estos registros de los del núcleo subtalámico.

```python
#Se agregan el nombre a la columnas
df_otras.columns = columnas

##Se agrega la nueva columna con la etiqueta
df_otras['estructura'] = 0

df_otras
```

### 3. Combinando los dos DataFrames previamente creados.

Uniendo los dos dataframes

```python
df_procesado = pd.concat([df_stn, df_otras])
df_procesado
```

Mezclando los registros

```python
df_procesado = df_procesado.sample(frac=1).reset_index(drop=True)
df_procesado
```

### 4. Guardando el nuevo dataframe

```python
df_procesado.to_csv('BD_procesada/BD_procesada.csv', index=False, sep=';')
```
